```python
#importing modules
import sys
from io import open
from IPython.display import Image, display
import time
from IPython.display import SVG, display
import os
!{sys.executable} -m pip install -q babachi==1.3.5
from babachi import BADEstimation
from babachi.visualize_segmentation import init_from_snps_collection
```

```python
# Datasets files
data_dir = './data'
badmaps_dir = './badmaps'
datasets = [os.path.join(data_dir, file) for file in os.listdir('./data')]
print(datasets)
```

```python
states = [1, 2, 3, 4, 5, 6, 4 / 3, 1.5, 2.5]
prior = dict(zip(states, [1] * len(states))) # dict with states as keys
chromosomes = ['chr2']
```

```python
for dataset_path in datasets:
    # Reading file
    with open(dataset_path, 'r') as dt:
        snps_collection, chr_ord, _ = BADEstimation.parse_input_file(dt)
        basename = os.path.splitext(os.path.basename(dataset_path))[0]
        badmap_file_path = os.path.join(badmaps_dir, basename + '.bed')
    

    GS = BADEstimation.GenomeSegmentator(
        snps_collection=snps_collection,
        chromosomes_order=chromosomes,
        out=badmap_file_path,
        states=states,
        prior=prior,
        verbose=True
    )
    try:
        GS.estimate_BAD()
    except Exception as e:
        raise e
    
    #visualizing part
    init_from_snps_collection(snps_collection, BAD_file=badmap_file_path, verbose=False, img_format='png')
    for chromosome in chromosomes:
        display(Image(os.path.join(badmaps_dir, '{}_visualization'.format(basename),
                                        '{}_{}.png'.format(basename, chromosome))))
    
    
# ROC 
```
